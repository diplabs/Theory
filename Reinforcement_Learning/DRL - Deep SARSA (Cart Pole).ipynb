```python
from keras.layers import Dense
from keras.models import Sequential
import numpy as np
import gym
from gym.envs.registration import register
from keras import optimizers
from keras import initializers
import random
```

```python
'''
환경 생성
'''
env = gym.make('CartPole-v1')
```

```python
n_1 = 12
n_2 = 16
n_3 = 6
```

```python
q = Sequential()
q.add(Dense(n_1, input_dim=env.observation_space.shape[0]))
q.add(Dense(n_2))
q.add(Dense(n_3))
q.add(Dense(env.action_space.n, activation='sigmoid'))
```

```python
q.compile(loss='mse', optimizer='adam')
```

```python
gamma = .95
epsilon = .2
episode = 0
max_episode = 1000

state = env.reset()
action = env.action_space.sample()
step = 0
total_r = 0

while(episode < max_episode):
    
    state = np.reshape(state, [1, 4])
    
    if(epsilon > 0):
        epsilon -= .0005
    
    # action 선택
    if(random.random() > epsilon):
        #print(pi)
        pi = q.predict(state)
        action = np.argmax(pi)
    else:
        action = env.action_space.sample()
        
    # step 진행
    state_next, reward, done, _ = env.step(action)
    state_next = np.reshape(state_next, [1, 4])

    step += 1
    total_r += reward

    # target 계산
    q_target = reward + gamma * q.predict(state_next)

    # episode 종료시
    if(done):
        q_target[0][action] = -100
        
        print(episode, total_r)
        step = 0
        episode += 1
        state_next = env.reset()
        total_r = 0
        
    q.fit(state, q_target, verbose=0)
    state = state_next

env.close()
```
